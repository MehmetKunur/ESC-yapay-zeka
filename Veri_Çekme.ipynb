```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.imdb.com/chart/top/"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.5" # Başlıkların İngilizce kalması için
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, "html.parser")

# Filmlerin bulunduğu liste öğeleri
movies = soup.find_all("li", class_="ipc-metadata-list-summary-item")

data = []

for movie in movies:
    try:
        # 1. Başlık (Title)
        title_tag = movie.find("h3", class_="ipc-title__text")
        title = title_tag.text.split(". ", 1)[1] if title_tag else "N/A"

        # 2. Yıl (Metadata) - Daha kararlı sınıf: 'cli-title-metadata-item'
        metadata = movie.find_all("span", class_="cli-title-metadata-item")
        year = metadata[0].text if metadata else "N/A"

        # 3. Puan (Rating)
        rating_tag = movie.find("span", class_="ipc-rating-star")
        rating = rating_tag.text.strip().split()[0] if rating_tag else "N/A"

        data.append({"Title": title, "Year": year, "Rating": rating})

    except (IndexError, AttributeError):
        # Hatalı satırı atla ama döngüyü kırma
        continue

# Verileri pandas DataFrame'e dönüştürme ve yazdırma
df = pd.DataFrame(data)
print(df.head())

# İstersen Excel veya CSV olarak kaydedebilirsin:
# df.to_excel("imdb_top_250.xlsx", index=False)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# 1. İstek Gönderme ve Kontrol
url = "https://www.imdb.com/chart/top/"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.5"
}

response = requests.get(url, headers=headers)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, "html.parser")

    # IMDb'nin güncel liste yapısı genellikle bu sınıftadır
    movies = soup.find_all("li", class_="ipc-metadata-list-summary-item")
    print(f"Bulunan film sayısı: {len(movies)}") # Kaç film yakalandığını görmeni sağlar

    data = []

    for movie in movies:
        try:
            # Başlık çekme
            title_tag = movie.find("h3", class_="ipc-title__text")
            # "1. The Shawshank Redemption" gibi gelen metinden sayıyı ayıklar
            title = title_tag.text.split(". ", 1)[1] if title_tag and ". " in title_tag.text else "N/A"

            # Yıl bilgisi çekme
            metadata = movie.find_all("span", class_="cli-title-metadata-item")
            year = metadata[0].text if len(metadata) > 0 else "N/A"

            # Puan bilgisi çekme
            rating_tag = movie.find("span", class_="ipc-rating-star")
            rating = rating_tag.text.strip().split()[0] if rating_tag else "N/A"

            data.append({"Title": title, "Year": year, "Rating": rating})
        except Exception as e:
            continue

    # 2. DataFrame Oluşturma
    df = pd.DataFrame(data)

    if not df.empty:
        print("Veri başarıyla çekildi!")
        display(df.head(10)) # İlk 10 filmi tablo olarak gösterir
    else:
        print("HATA: Veri çekilemedi, liste boş. IMDb sınıf isimlerini güncellemiş olabilir.")
else:
    print(f"HATA: Sayfaya erişilemedi. Durum kodu: {response.status_code}")

# Clipboard yerine doğrudan 'df' değişkenini kullanabilirsin
```
